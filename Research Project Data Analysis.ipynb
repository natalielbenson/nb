```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns

%matplotlib inline
```

## Length and width of vessels in 2D culture

```python
df1 = pd.read_excel('cdx17.xlsx')
df1.head()
```

```python
df1.describe()
```

```python
df2 = pd.read_excel('cdx19.xlsx')
df2.head()
```

```python
df2.describe()
```

```python
df3 = pd.read_excel('huvec.xlsx')
df3.head()
```

```python
df3.describe()
```

```python
data1 = pd.DataFrame(df1, columns=['Cultrex','Geltrex']).assign(Location=1)
data2 = pd.DataFrame(df2+0.2, columns=['Cultrex','Geltrex']).assign(Location=2)
data3 = pd.DataFrame(df3+0.4, columns=['Cultrex','Geltrex']).assign(Location=3)

cdf = pd.concat([data1, data2, data3])    
mdf = pd.melt(cdf, id_vars=['Location'], var_name=['Hydrogel'])
print(mdf.head())

ax = sns.boxplot(x="Location", y="value", hue="Hydrogel", data=mdf, palette = ['plum', 'mediumseagreen']) 
ax.set_xticks([0,1,2])
ax.set_xticklabels(['CDX17', 'CDX19', 'HUVEC'])

plt.xlabel('Cell Type')
plt.ylabel('Length (µm)')

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

sns.despine(top=True,
            right=True,
            left=False,
            bottom=False)

plt.savefig('length.png', dpi =400, bbox_inches = "tight")

plt.show()
```
